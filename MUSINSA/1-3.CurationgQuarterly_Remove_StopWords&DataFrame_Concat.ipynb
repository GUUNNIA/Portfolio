```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

from collections import Counter
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
def musinsa_words(words_list):
  word_list = []
  for goods in words_list:
    words = goods.split(',')
    for word in words:
      word_list.append(word)
  return word_list

def drop_words(word_list):
  word_list_drop = []
  for word in word_list:
    if len(word) != 1:   # 한글자 단어는 Drop 
      word_list_drop.append(word)

  return word_list_drop
```

```python
season_one = pd.read_csv('/content/2021_1분기.csv')
season_two = pd.read_csv('/content/2021_2분기.csv')
season_three = pd.read_csv('/content/2021_3분기.csv')
season_four = pd.read_csv('/content/2021_4분기.csv')
```

```python
season_one
```

```python
season_one_list = musinsa_words(season_one['words'])
season_two_list = musinsa_words(season_two['words'])
season_three_list = musinsa_words(season_three['words'])
season_four_list = musinsa_words(season_four['words'])
```

```python
season_one_list
```

```python
season_one_list = drop_words(season_one_list)
season_two_list = drop_words(season_two_list)
season_three_list = drop_words(season_three_list)
season_four_list = drop_words(season_four_list)
```

```python
counter_1 = dict(Counter(season_one_list))
counter_2 = dict(Counter(season_two_list))
counter_3 = dict(Counter(season_three_list))
counter_4 = dict(Counter(season_four_list))
```

```python
counter_1
```

```python
def to_df(data):
  key_data = list(data.keys())
  values_data = list(data.values())

  word_dic = {
      'word' : [],
      'conunt' : []
  }

  for key, values in zip(key_data, values_data):
    word_dic['word'].append(key)
    word_dic['conunt'].append(values)

  word_df = pd.DataFrame(word_dic)
  
  return word_df
```

```python
counter_1_df = to_df(counter_1)
counter_2_df = to_df(counter_2)
counter_3_df = to_df(counter_3)
counter_4_df = to_df(counter_4)
```

```python
counter_1_df
```

```python
one_sort_100 = counter_1_df.sort_values('conunt', ascending=False)[:100].reset_index().drop('index', axis=1)
two_sort_100 = counter_2_df.sort_values('conunt', ascending=False)[:100].reset_index().drop('index', axis=1)
three_sort_100 = counter_3_df.sort_values('conunt', ascending=False)[:100].reset_index().drop('index', axis=1)
four_sort_100 = counter_4_df.sort_values('conunt', ascending=False)[:100].reset_index().drop('index', axis=1)
```

```python
one_sort_100.columns = ['word_1', 'count_1']
two_sort_100.columns = ['word_2', 'count_2']
three_sort_100.columns = ['word_3', 'count_3']
four_sort_100.columns = ['word_4', 'count_4']
```

```python
all_data = pd.concat([one_sort_100, two_sort_100, three_sort_100, four_sort_100], axis=1)
```

<h4> 불용어 제거 </h4>

```python
stop_words = '아이템, 컬러, 디자인, 매력, 하나, 코디, 완성, 구성, 로고, 무드, 연출, 소재, 추천, 느낌, 착용, 덕분, 일리, 아우, 오버, 감성, 스타, 사이즈, 원단, 일링, 바로, 시즌, 선택, 가지, 분위기, 선사, 준비, 가장, 더욱, 자랑, 지금, 때문, 미니, 서머, 여유, 또한, 마음, 부담, 피부, 라인, 여유, 매치, 제격, 촉감, 자리, 길이, 날씨, 적용, 부분, 취향, 이번, 미니, 고민, 생각, 일이, 특징, 테일로, 누구, 여기, 시선, 군더더기, 기능, 부분, 자리, 밑단, 가장, 웨어'
stop_words =  stop_words.split(', ')
```

```python
def del_stop(data):
  words = []
  for w in data:
    if w not in stop_words:
      words.append(w)
  return words
```

```python
one_quarter = del_stop(all_data['word_1'])
two_quarter = del_stop(all_data['word_2'])
three_quarter = del_stop(all_data['word_3'])
four_quarter = del_stop(all_data['word_4'])
```

```python
one_quarter_df = pd.DataFrame(one_quarter) 
two_quarter_df = pd.DataFrame(two_quarter)
three_quarter_df = pd.DataFrame(three_quarter)
four_quarter_df = pd.DataFrame(four_quarter)
```

```python
one_quarter_df.columns = ['1_quarter']
two_quarter_df.columns = ['2_quarter']
three_quarter_df.columns = ['3_quarter']
four_quarter_df.columns = ['4_quarter']
```

```python
all_quarter = pd.concat([one_quarter_df, two_quarter_df, three_quarter_df, four_quarter_df], axis=1)
```
